# 6. Unpack the rating data:

```python
ted.ratings.head()
```

```python
ted.ratings[0]
```

```python
def get_num_ratings (list_of_dicts):
    num = 0
    for d in list_of_dicts:
        num = num + d['count']
    return num
```

```python
get_num_ratings(ted.ratings_list[0])
```

```python
ted.ratings_list.apply(get_num_ratings)
```

```python
pd.DataFrame(ted.ratings_list[0])['count'].sum()
```

# Which occupation deliver the funniest TED talks on average?

```python
ted.ratings.str.contains('Funny').value_counts()
```

## 1. step one

```python
def get_funny_ratings(list_of_dicts):
    for d in list_of_dicts:
        if d['name'] == 'Funny':
            return d['count']
```

```python
ted.ratings_list.apply(get_funny_ratings)
```
